```python
# CONTROL FLOW IN PYTHON

# 1. IF STATEMENT
age = 20

if age >= 18:
    print("You are an adult")

# 2. IF-ELSE STATEMENT
number = 7

if number % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")

# 3. IF-ELIF-ELSE STATEMENT
score = 85

if score >= 90:
    print("Grade A")
elif score >= 80:
    print("Grade B")
elif score >= 70:
    print("Grade C")
else:
    print("Grade D")

# 4. FOR LOOP
print("\nFor Loop:")
for i in range(1, 6):
    print(i)

# 5. WHILE LOOP
print("\nWhile Loop:")
count = 1

while count <= 5:
    print(count)
    count += 1

# 6. BREAK
print("\nBreak Example:")
for i in range(10):
    if i == 5:
        break
    print(i)

# 7. CONTINUE
print("\nContinue Example:")
for i in range(10):
    if i == 5:
        continue
    print(i)

# 8. PASS
print("\nPass Example:")
for i in range(3):
    if i == 1:
        pass
    print(i)
```
